```python
df=df.drop_duplicates()
```

```python
df['Gender']=df['Gender'].astype('category')
```

```python
df['Workout_Type']=df['Workout_Type'].astype('category')
```

```python
df['HRR']=df['Max_BPM']-df['Resting_BPM']
```

```python
def BMI(bc):
  if bc<18.5 :return 'Under Weight'
  elif bc<25 :return 'Normal Weight'
  elif bc<30 :return 'Over Weight'
  else:return 'Obese'
```

```python
df['BMI_Category']=df['BMI'].apply(BMI)
```

```python
df['BMI_Checking']=df['Weight (kg)']/(df['Height (m)']**2)
```

```python
df['Calories_per_Hour'] = df['Calories_Burned'] / df['Session_Duration (hours)']
```

```python
grp=df.groupby('Workout_Type',as_index=False)['Calories_Burned'].count().sort_values(by=['Calories_Burned'],ascending=False)
```

Below the chart is trying to show what workout_type is most effective in terms of burning calories.


```python
px.bar(grp,x='Workout_Type',y='Calories_Burned',title='Effective_Workout',color='Workout_Type').show()
```

Below this pie chart we are trying to show which  type of workout is effective. which is show in the form of percentage.

```python
px.pie(grp,names='Workout_Type',values='Calories_Burned',title='Effective_Workout').show()
```

```python
df['Age']=df['Age'].astype('int')
```

```python
df.info()
```

```python
def age_group(ag):
    if 20 <= ag <= 30:
        return 'Young'
    elif 31 <= ag <= 40:
        return 'Adult'
    elif 41 <= ag <= 50:
        return 'Middle_aged'
    else:
        return 'Senior'
df['age_group']=df['Age'].apply(age_group)
```

```python
df.columns
```

```python
grp=df.groupby('age_group',as_index=False)['Experience_Level'].count().sort_values(by=['Experience_Level'],ascending=False)
```

 Below in this pie chart we are trying to show which age group has the most and least experience level.

```python
px.pie(grp,names='age_group',values='Experience_Level',title='Age_Based_Experience_Level').show()
```

Below is an histogram in this graph we are trying to show the distribution of gym members in terms of their BMI_Category.

```python
px.histogram(df,x='BMI_Category',title='Categories').show()
```

```python
grp=df.groupby('Age',as_index=False)['BMI'].count().sort_values(by=['BMI'],ascending=False)
```

This scatter plot shows the relationship between Age (x-axis) and BMI (y-axis).

```python
px.scatter(df,x='Age',y='BMI',title='Age v/s BMI').show()
```

```python
grp=df.groupby('Gender',as_index=False)['Workout_Frequency (days/week)'].count().sort_values(by=['Workout_Frequency (days/week)'],ascending=False)
```

This bar graph shows which gender(MALE/FEMALE) and their Workout Frequency (No.of days or weeks working out).
